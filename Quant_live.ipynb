```python
from Quant_live import Strategy
import json
import pandas as pd
import plotly.graph_objects as go

def run_simulation(json_file: str):
    strategy = Strategy()
    with open(json_file, "r") as f:
        events = json.load(f)
    for event in events:
        strategy.on_game_event_update(**event)

if __name__ == "__main__":
    run_simulation("example-game.json")
```

```python

df_raw = pd.read_json("example-game.json")
df_raw['time_seconds']=df_raw['time_seconds'].sort_values(ascending=True).reset_index(drop=True)
time_col = "time_seconds"
score_cols = ["home_score", "away_score"]
meta_cols = [c for c in df_raw.columns if c not in {time_col, *score_cols}]

customdata = df_raw[meta_cols].fillna("N/A").to_numpy()

def hover(label):
    lines = "<br>".join(f"{col}: %{{customdata[{i}]}}" for i, col in enumerate(meta_cols))
    return f"<b>{label} score:</b> %{{y}}<br>{lines}<extra></extra>"

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df_raw[time_col],
    y=df_raw["home_score"],
    mode="lines",
    name="home_score",
    customdata=customdata,
    hovertemplate=hover("Home"),
    
))
fig.add_trace(go.Scatter(
    x=df_raw[time_col],
    y=df_raw["away_score"],
    mode="lines",
    name="away_score",
    customdata=customdata,
    hovertemplate=hover("Away"),
))

fig.update_layout(
    title="Home vs Away Score Over time_seconds",
    xaxis_title=time_col,
    yaxis_title="Score",
)

fig.show()
```
